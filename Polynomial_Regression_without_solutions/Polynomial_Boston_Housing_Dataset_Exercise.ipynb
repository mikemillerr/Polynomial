```python
corr_treshold = 0.68
df_corr = corrmat.sort_values(by='medv', axis=0, ascending=False)
# Create a list of the features above corr_treshold
high_corr_feat = list(df_corr['medv'][df_corr['medv'].abs() > corr_treshold].index)
if 'medv' in high_corr_feat: high_corr_feat.remove('medv')
high_corr_feat
```

## Scaling and Splitting 

```python
test_size = 0.2
df_mlr = df.loc[:, high_corr_feat]

# transform dataframe to numpy array
X = df_mlr.to_numpy()
y = df['medv'].to_numpy()


if scaling:
    # Scaling x
    #scaler = MinMaxScaler()
    scaler = StandardScaler()
    scaled_x = scaler.fit_transform(df.iloc[:,:-1])
    columns = df.columns
    X = pd.DataFrame(scaled_x, columns=columns[:-1])
    
# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=5)
```

## Multiple Linear Regression with all high_corr_features

```python
# Reshaping the arrays
y_train = y_train[:, np.newaxis]
y_test = y_test[:, np.newaxis]

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(X_train, y_train)

# Make predictions using the testing set
y_pred_test = regr.predict(X_test)
y_pred_train = regr.predict(X_train)

print('THIS ARE THE RESULTS FOR MULTIPLE LINEAR REGRESSION:')
# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred_test))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, y_pred_test))
# The root mean squared error
print("Mean squared error: %.2f" % np.sqrt(mean_squared_error(y_test, y_pred_test)))
```

## Polynomial Regression with all high_corr_feat

```python
mse_list = []
rmse_list = []
r2_list = []


for i in range(1,8):
    polynomial_features= PolynomialFeatures(degree=i)
    X_poly_train = polynomial_features.fit_transform(X_train)
    X_poly_test = polynomial_features.fit_transform(X_test)

    # Create linear regression object
    regr_poly = linear_model.LinearRegression()

    # Train the model using the training sets
    regr.fit(X_poly_train, y_train)

    # Make predictions using the testing set
    y_pred_poly_test = regr.predict(X_poly_test)
    y_pred_poly_train = regr.predict(X_poly_train)
    
    mse = mean_squared_error(y_test, y_pred_poly_test)
    r2 = r2_score(y_test, y_pred_poly_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred_poly_test))
    
    mse_list.append(mse)
    r2_list.append(r2)
    rmse_list.append(rmse)
    
    print('THIS ARE THE RESULTS FOR POLYNOMIAL REGRESSION GRADE %s:' %i)
    # The mean squared error
    print("Mean squared error: %.2f" % mse)
    # Explained variance score: 1 is perfect prediction
    print('Variance score: %.2f' % r2)
    # The root mean squared error
    print("Root Mean squared error: %.2f" % rmse)
    
    
```
